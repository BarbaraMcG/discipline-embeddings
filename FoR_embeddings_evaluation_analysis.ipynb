## All parameter values

# Functions

Function for defining first and last year


```
# Function that defines first and last year for embeddings:
def define_first_last_year(year_intervals):
  last_year = 2019
  if year_intervals == 3:
    first_year = 1990 
  elif year_intervals == 5:
    first_year = 1950
  return (first_year, last_year)
```

Function that defines evaluation file names


```
def define_eval_file_summary_name(year_intervals, num_dimensions, window_size, negative_sampling):
  eval_file_summary_name = BASE_DIR + '/analysis/evaluation/Evaluation_yearint-' + str(year_intervals) + '_numdim-' + str(num_dimensions) + '_window-' + str(window_size) + '_negsamp-' + str(negative_sampling) + '.txt'
  return eval_file_summary_name
```

# Read FoR embeddings evaluation files

Choose the best hyperparameters for the FoR embeddings 



```
average_successes = dict() #maps a hyperparameter tuple to the average of the number of successes in the evaluation file for that tuple 
```

Read files that contain evaluation metrics for FoR embeddings

```
for year_intervals in all_year_intervals:
  print(str(year_intervals), "year intervals")
  for num_dimensions in all_num_dimensions:
    print(str(num_dimensions), "dimensions")
    for window_size in all_window_sizes:
      print(str(window_size), "window size")
      for negative_sampling in all_negative_sampling:
        print(str(negative_sampling), "negative sampling")
        #print("Reading file", define_eval_file_summary_name(year_intervals, num_dimensions, window_size, negative_sampling))
        eval_df = pd.read_csv(define_eval_file_summary_name(year_intervals, num_dimensions, window_size, negative_sampling), sep='\t', header = 1)
        #print(str(eval_df))
        average_successes[window_size, negative_sampling] = eval_df.iloc[:,1].mean()
```

```
average_successes_df = pd.DataFrame.from_dict(average_successes, orient='index')
average_successes_df
```
